```python
x, s = sympy.symbols('x s')
f = x**2
slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)
result = sympy.limit(slope_f, s, 0)
result
```

# The Chain Rule

```python
x, y = sympy.symbols('x y')
_y = x**2 + 1
dy_dx = sympy.diff(_y)
z = y**3 - 2
dz_dy = sympy.diff(z)
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = sympy.diff(z.subs(y, _y))
```

```python
dz_dx_chain
```

```python
dz_dx_no_chain
```

# Integrals

```python
x = sympy.symbols('x')
f = x**2 + 1
area = sympy.integrate(f, (x, 0, 1))
area
```

```python
x, i, n = sympy.symbols('x i n')
f = x**2 + 1
lower, upper = 0, 1
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)
n_rectangles = sympy.Sum(delta_x * fx_i, (i, 1, n)).doit()
area = sympy.limit(n_rectangles, n, "oo")
area
```
